```python
%pylab inline
import sys
sys.path.append('./JSAnimation')
from JSAnimation import IPython_display
from matplotlib import animation
from microTDFL import *
```

```python
n = 40 # number of charge segments (one greater than number of current segments)
dx = 5
dt = 35e-9
a = 0.003
R = 0 * dx # resistance between adjacent charges
E_app = 100000
navg = 501

m = timeStepsNeeded(n, dx, dt)
R2 = computeR2(n, dx, a, navg)
R1L = computeR1L(n, dx, a, navg)
R1T = computeR1T(n, dx, a, navg)
#G = getTransitionMatrix(n, m, dx, dt, E_app, R, R2, R1L, R1T)
G = getTwoStageMatrix(n, m, dx, dt, E_app, R, R2, R1L, R1T)
x = linalg.eig(G)
```

```python
print max(abs(x[0]))
plot(real(x[0]), imag(x[0]), 'ro')
plot(cos(linspace(0, 2*pi)), sin(linspace(0, 2*pi)), 'r')
show()
plot(angle(x[0]) / dt, -log(abs(x[0])) / dt, 'bo')
show()
```

```python
plot(real(x[1][m*(n-1):-1,820]))
show()
```

```python
steps = 150
h = zeros((m * (n - 1) + n + 1, steps))
h[-1, 0] = 0
h[-2, 0] = 1
for i in range(1, steps):
    h[:,i] = G.dot(h[:,i-1])
```

```python
fig = plt.figure()
#ax = plt.axes(xlim=(0, n - 1), ylim=(amin(h[m*(n-1):m*(n-1)+n,:]), amax(h[m*(n-1):m*(n-1)+n,:])))
ax = plt.axes(xlim=(0, n - 1), ylim=(amin(h[:n-1,:]), amax(h[:n-1,:])))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    #line.set_data(range(n), h[m*(n-1):m*(n-1)+n,i])
    line.set_data(range(n-1), h[:n-1,i])
    return line,

animation.FuncAnimation(fig, animate, init_func=init, frames=steps, interval=20, blit=True)
```
